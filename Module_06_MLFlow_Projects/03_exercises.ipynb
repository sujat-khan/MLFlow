## Exercise 1: Run the Sample Project with Different Configs

Use `mlflow.projects.run()` to run the sample project with these 3 configurations:
1. `n_estimators=50, max_depth=3`
2. `n_estimators=150, max_depth=7`
3. `n_estimators=300, max_depth=15`

Then use `mlflow.search_runs()` to find the best configuration.

```python
# YOUR CODE HERE
```

```python
# ✅ SOLUTION
project_dir = "sample_project"
if not os.path.exists(project_dir):
    project_dir = os.path.join(os.path.dirname(os.getcwd()), "Module_06_MLFlow_Projects", "sample_project")

configs = [
    {"n_estimators": 50, "max_depth": 3},
    {"n_estimators": 150, "max_depth": 7},
    {"n_estimators": 300, "max_depth": 15},
]

run_ids = []
for cfg in configs:
    result = mlflow.projects.run(
        uri=project_dir,
        entry_point="train",
        env_manager="local",
        parameters=cfg,
    )
    run_ids.append(result.run_id)
    print(f"  Config {cfg} → Run {result.run_id[:8]}...")

# Find best
runs = mlflow.search_runs(
    filter_string=" OR ".join([f"run_id = '{r}'" for r in run_ids]),
    order_by=["metrics.accuracy DESC"]
)
cols = [c for c in ["params.n_estimators", "params.max_depth", "metrics.accuracy"] if c in runs.columns]
if cols:
    print(f"\n🏆 Results:")
    print(runs[cols].to_string(index=False))
```
